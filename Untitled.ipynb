```python
import cv2
```

```python
def send_frame(camera):
    x = []
    for i in range(1,3):
        retval,im = camera.read()
        imgencode = cv2.imencode('.jpg',im)[1]
        stringData=imgencode.tostring()
        x.append(stringData)
    return x
```

```python
camera=cv2.VideoCapture(0)
t = str({'data': send_frame(camera)})
print(t)
camera.release()
```
